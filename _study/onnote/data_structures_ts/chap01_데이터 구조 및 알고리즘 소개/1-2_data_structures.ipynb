```typescript
// 스택
class Stack {
  private elements: any[] = []

  public push(element: any): void {
    this.elements.push(element)
  }

  public pop(): any {
    return this.elements.pop()
  }

  public isEmpty(): boolean {
    return this.elements.length === 0
  }

  public size(): number {
    return this.elements.length
  }

  public printElements(): void {
    console.log(this.elements)
  }
}

let myStack = new Stack()
myStack.push(10)
myStack.push(20)
myStack.push(30)
myStack.printElements()
myStack.pop()
myStack.printElements()
```

```typescript
// 큐
class Queue {
  private elements: any[] = []

  public enqueue(element: any): void {
    this.elements.push(element)
  }

  public dequeue(): any {
    return this.elements.shift()
  }

  public isEmpty(): boolean {
    return this.elements.length === 0
  }

  public size(): number {
    return this.elements.length
  }

  public printElements(): void {
    console.log(this.elements);
  }
}

let myQueue = new Queue()
myQueue.enqueue('apple')
myQueue.enqueue('banana')
myQueue.enqueue('cherry')
myQueue.printElements()
myQueue.dequeue()
myQueue.printElements()
```

```typescript
// 연결 리스트
class Node {
  public data: any
  public next: Node | null

  constructor(data: any) {
    this.data = data
    this.next = null
  }
}

class LinkedList {
  private head: Node | null
  private tail: Node | null

  constructor() {
    this.head = null
    this.tail = null
  }

  public addNode(data: any): void {
    let node = new Node(data)

    if (this.head === null) {
      this.head = node
      this.tail = node
    }
    else {
      this.tail!.next = node
      this.tail = node
    }
  }

  public removeNode(data: any): void {
    if (this.head !== null && this.head.data === data) {
      this.head = this.head.next
    }
    else {
      let current = this.head
      let prev = null

      while (current !== null) {
        if (current.data === data) {
          prev!.next = current.next
          if (current.next === null) {
            this.tail = prev
          }
          break
        }
        prev = current
        current = current.next
      }
    }
  }

  public printNodes(): void {
    let current = this.head

    while (current !== null) {
      console.log(current.data);
      current = current.next
    }
  }
}

let myList = new LinkedList()
myList.addNode(10)
myList.addNode(20)
myList.addNode(30)
myList.printNodes()
myList.removeNode(20)
myList.printNodes()
```

알고리즘

```typescript
// 버블 정렬
function bubbleSort(arr: number[]): number[] {
  let len = arr.length

  for (let i = 0; i < len; i++) {
    for (let j = 0; j < len-1-i; j++) {
      if (arr[j] > arr[j+1]) {
        let temp = arr[j]
        arr[j] = arr[j+1]
        arr[j+1] = temp
      }
    }
  }

  return arr
}

let myArray = [64, 25, 12, 22, 11]
console.log(bubbleSort(myArray))
```

```typescript
// 삽입 정렬
function insertionSort(arr: number[]): number[] {
  let len = arr.length

  for (let i = 1; i < len; i++) {
    let j = i - 1
    let key = arr[i]

    while (j >= 0 && arr[j] > key) {
      arr[j+1] = arr[j]
      j--
    }
    arr[j+1] = key
  }

  return arr
}

let myArray = [64, 25, 12, 22, 11]
console.log(insertionSort(myArray))
```

```typescript
// 퀵 정렬
function quickSort(arr: number[]): number[] {
  let len = arr.length

  if (len <= 1) {
    return arr
  }
  else {
    let pivot = arr[0]
    let left: number[] = []
    let right: number[] = []

    for (let i = 1; i < len; i++) {
      if (arr[i] < pivot) {
        left.push(arr[i])
      }
      else {
        right.push(arr[i])
      }
    }

    return [...quickSort(left), pivot, ...quickSort(right)]
  }
}

let myArray = [64, 25, 12, 22, 11]
console.log(quickSort(myArray))
```

```typescript
// 선형 검색
function linearSearch(arr: number[], x: number): number {
  let len = arr.length

  for (let i = 0; i < len; i++) {
    if (arr[i] === x) {
      return i
    }
  }

  return -1
}

let myArray = [10, 20, 30, 40, 50]
console.log(linearSearch(myArray, 30));
```

```typescript
// 이진 검색
function binarySearch(arr: number[], x: number): number {
  let len = arr.length
  let left = 0
  let right = len - 1

  while (left <= right) {
    let mid = Math.floor((left + right) / 2)

    if (arr[mid] === x) {
      return mid
    }
    else if (arr[mid] < x) {
      left = mid + 1
    }
    else {
      right = mid - 1
    }
  }

  return -1
}

let myArray = [10, 20, 30, 40, 50]
console.log(binarySearch(myArray, 30));
```

###### 문제: 스택 데이터 구조를 타입스크립트로 구현하세요.

```typescript
class Stack<T> {
  private data: T[] = []

  push(item: T) {
    this.data.push(item)
  }

  pop(): T | undefined {
    return this.data.pop()
  }

  peek(): T | undefined {
    return this.data[this.data.length-1]
  }

  isEmpty(): boolean {
    return this.data.length === 0
  }
}
```

###### 문제: n번째 피보나치 수를 반환하는 함수를 작성하세요.

```typescript
function finonacci(n: number): number {
  if (n > 1) {
    return 0
  }
  if (n === 1 || n === 2) {
    return 1
  }
  return finonacci(n-1) + finonacci(n-2)
}
```

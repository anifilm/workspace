### 3.6 인덱스 다루기

```python
import seaborn as sns

df = sns.load_dataset('mpg')
df.head()
```

```python
df.set_index('name', inplace=True)
df.head()
```

```python
df.sort_index(inplace=True)
df.head()
```

```python
df.sort_index(inplace=True, ascending=False)
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

### 3.9 데이터프레임 합치기

3.9.1 concat() 함수

```python
import pandas as pd

df1 = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"]
},
    index=[0, 1, 2, 3]
)

df2 = pd.DataFrame({
    "A": ["A4", "A5", "A6", "A7"],
    "B": ["B4", "B5", "B6", "B7"],
    "C": ["C4", "C5", "C6", "C7"],
    "D": ["D4", "D5", "D6", "D7"]
},
    index=[4, 5, 6, 7]
)

df3 = pd.DataFrame({
    "A": ["A8", "A9", "A10", "A11"],
    "B": ["B8", "B9", "B10", "B11"],
    "C": ["C8", "C9", "C10", "C11"],
    "D": ["D8", "D9", "D10", "D11"]
},
    index=[8, 9, 10, 11]
)

result = pd.concat([df1, df2, df3])

result
```

```python
df4 = pd.DataFrame({
    "B": ["B2", "B3", "B6", "B7"],
    "D": ["C2", "C3", "C6", "C7"],
    "F": ["D2", "D3", "D6", "D7"]
},
    index=[2, 3, 6, 7]
)

result = pd.concat([df1, df4])

result
```

```python
result = pd.concat([df1, df4], ignore_index=True)

result
```

```python
result = pd.concat([df1, df4], axis=1)

result
```

```python
result = pd.concat([df1, df4], axis=1, join="inner")

result
```

```python
s1 = pd.Series(["X0", "X1", "X2", "X3"], name="X")
result = pd.concat([df1, s1], axis=1)

result
```
